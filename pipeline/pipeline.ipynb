# 🔄 Pipeline ETL
## Nettoyage et transformation

```python
import sys
sys.path.append('..')
  
import pandas as pd
import glob
import os
from pipeline.cleaner import DataCleaner
from pipeline.transformer import DataTransformer
    
print("✅ Modules importés")
```

```python
# Charger le dernier fichier brut

csv_files = glob.glob('../data/raw/jobs_*.csv')
if csv_files:
    latest_file = max(csv_files, key=os.path.getctime)
    df_raw = pd.read_csv(latest_file)
    print(f"📁 Fichier chargé: {latest_file}")
    print(f"📊 Dimensions: {df_raw.shape}")
else:
    print("❌ Aucun fichier trouvé")
```

```python
# Nettoyage

cleaner = DataCleaner()
df_clean = cleaner.clean(df_raw)
df_clean.head()
```

```python
# Transformation

transformer = DataTransformer()
df_trans = transformer.transform(df_clean)
df_trans.head()
```

```python
# Sauvegarde

df_trans.to_csv('../data/processed/jobs_clean.csv', index=False)
print(f"✅ Données sauvegardées dans data/processed/jobs_clean.csv")
```
